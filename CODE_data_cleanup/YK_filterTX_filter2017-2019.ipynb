```python
import pandas as pd
import time
```

```python
# import traffic data from original CSV
path="data/US_Accidents_June20.csv"
traffic_df = pd.read_csv(path)
```

```python
traffic_df.head() # debug print
```

```python
traffic_df.columns # debug print
```

```python
traffic_df['Year']=traffic_df['Start_Time'].str.split("-", n = 1, expand = True)[0]
```

```python
traffic_df['Year'].isnull().values.any()
```

```python
city_groupby=traffic_df.groupby(['State','County','City'])
```

```python
city_groupby.agg({'ID':'count'}).sort_values("ID",ascending=False)
```

```python
# filter only TX data
start = time.time()
texas_df=traffic_df.loc[traffic_df['State']=='TX'].copy()
# Split Start column to get year and  Create Year Column in data
texas_df['Year']=texas_df['Start_Time'].str.split("-", n = 1, expand = True)[0]
#year=year_time_df[0] # debug Print
end = time.time()
print(end - start)
```

```python
# Create Year Column in data
texas_df.head() # Debug print
```

```python
# filter only 2019-2017 years
texas_df=texas_df.loc[(texas_df['Year']=='2019')| (texas_df['Year']=='2018')|(texas_df['Year']=='2017')].reset_index(drop=True)
```

```python
#write to csv
texas_df.to_csv('data/YK_TX2017-19_Accidents.csv', index=False)
```
